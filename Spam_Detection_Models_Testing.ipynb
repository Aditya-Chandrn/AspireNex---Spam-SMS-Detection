```python
import joblib
import pandas as pd
import os
import random
```

### Create test SMS

```python
test_sms = {
    "tag": ["spam1", "spam2", "spam3", "ham1", "ham2", "ham3"],

    "text": [
        "Congratulations! You have been selected as a winner of a $1000 cash prize! Text WIN to 12345 to claim your prize now.",
        "Need cash fast? Get a quick loan of up to $5000 with no credit check! Apply now at www.quickloan.com. Hurry, offer ends soon!",
        "You've been chosen for an exclusive offer! Get a brand new iPhone for just $1. Click here to claim: www.cheapiphones.com.",

        "Reminder: Your appointment with Dr. Smith is scheduled for tomorrow at 10:00 AM. Please arrive 15 minutes early. Thank you!",
        "Hey! Are we still on for dinner at 7 PM tonight? Let's meet at the new Italian place downtown. Looking forward to it!",
        "Good morning team, there's a meeting at 2 PM today in the conference room to discuss the new project. Please be on time."
    ]
}
```

```python
# shuffle the dataset
df = pd.DataFrame(test_sms)
df = df.sample(frac=1).reset_index(drop=True)
df
```

### Show comparison between expectation and prediction

```python
def display_result(model_name, prediction):
    print("\n\n", model_name)
    result_df = pd.DataFrame()
    result_df["tag"] = df["tag"]
    result_df["prediction"] = ["spam" if pred == 1 else "ham" for pred in prediction]

    display(result_df)
```

### Load all models and predict spam SMS

```python
dir = "/content/saved_models"
saved_models = os.listdir(dir)

models = []

for model_name in saved_models:
    model = joblib.load(os.path.join(dir, model_name))
    name = model_name.split(".")[0]
    models.append(name)
    prediction = model.predict(df["text"])
    display_result(name, prediction)
```
